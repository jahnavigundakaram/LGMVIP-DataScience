# Importing Library

```python
import cv2
```

# Reading Image

```python
image = cv2.imread("Jerry.jpg")
```

# Displaying Image

```python
cv2.imshow("Original Image",image)
cv2.waitKey(0)
```

# Reading Image in RGB Format

```python
import matplotlib.pyplot as plt 
rgb_img = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
rgb_img
```

# Converting Original Image to Graysacle

```python
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_img
```

```python
plt.imshow(gray_img)
```

# Inverting Image

```python
inverted_img = 255 - gray_img
plt.imshow(inverted_img)
```

# Blur Image usung Gauss Function

```python
blur_img = cv2.GaussianBlur(inverted_img,(21,21),0)
plt.imshow(blur_img)
```

# Pencil Sketch

```python
inverted_blur = 255 - blur_img
pencil_sketch = cv2.divide(gray_img, inverted_blur, scale = 256.0)
plt.imshow(pencil_sketch)
```

```python
cv2.imshow("Original Image", pencil_sketch)
cv2.waitKey(0)
```
